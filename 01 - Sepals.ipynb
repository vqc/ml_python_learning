```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0)
```

```python
#visualize the test data

fig, ax = plt.subplots(3, 3, figsize=(15, 15))
plt.suptitle("iris_pairplot")

for i in range(3):
    for j in range(3):
        ax[i, j].scatter(X_train[:, j], X_train[:, i + 1], c=y_train, s=60)
        ax[i, j].set_xticks(())
        ax[i, j].set_yticks(())
        if i==2:
            ax[i, j].set_xlabel(iris['feature_names'][j])
        if j==0:
            ax[i, j].set_ylabel(iris['feature_names'][i+1])
        if j>i:
            ax[i, j].set_visible(False)
            
        
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors =1)
```

```python
#train the classifier
knn.fit(X_train, y_train)
```

```python
#generate prediction using test data using the trained classifier
y_pred = knn.predict(X_test)
#compare with the known outcomes
np.mean(y_pred == y_test)
#answer is accuracy => how close it is to true. precision is a measure of consistency
```

```python
#do previous comparison using built in function
knn.score(X_test, y_test)
```
